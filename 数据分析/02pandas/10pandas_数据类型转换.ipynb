```python
import pandas as pd
data = pd.DataFrame({
    'name':['alice','alice','bob','alice','jack','bob'],
    'age':[26,25,30,25,35,30],
    'city':['NY','NY','LA','NY','SF','LA']
})
data
```

```python
data.duplicated()  # 一整条记录都是一样的，标记为重复，返回True
data.drop_duplicates()  # 去重
data.drop_duplicates(subset=['name'])  # 指定某一列重复的去重,留下最前面的
data.drop_duplicates(subset=['name'],keep='last') # 指定某一列重复的去重,留下最后面的
```
