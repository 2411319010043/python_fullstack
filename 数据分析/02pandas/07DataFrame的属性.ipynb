```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'id':[1,2,3,4,5],
    'name':['tom','jack','alice','luce','mike'],
    'age':[15,17,20,26,30],
    'score':[23,45,78,99,44]
},index=[1,2,3,4,5],columns=['name','id','age','score']  # columns调整列的顺序
)
```

```python
# 1. index
print('行索引:',df.index)
```

```python
# 2. 列标签
print('列标签:',df.columns)
```

```python
# 3. 值
print('值:',df.values)
```

```python
# 4. ndim 维度
print('维度:',df.ndim)
```

```python
# 4. dtypes 数据类型
print('数据类型:',df.dtypes)
```

```python
# 5. shape 形状
print('形状:',df.shape)
```

```python
# 6. size 元素个数
print('元素个数:',df.size)
```

```python
# 7. 行列转置
print(df.T)
```

```python
print(df)
```

```python
# 8. 获取元素 loc iloc at iat
# 获取某行的数据
print(df.loc[4])
print(df.iloc[3])
# 获取某列数据
print(df.loc[:,'name'])
print(df.iloc[:,0])
print(df['name'])
print(df.name)
print(df[['name']])
```

```python
# 获取单个数据
print(df.at[3,'name'])
print(df.iat[2,0])
```

```python
# 查看部分数据
print(df.head())
print(df.tail(3))
```

```python
# 使用布尔索引筛选数据
df[df.score>70]
df[ (df['score']>70) & (df.age<20)]
```

```python
# 随机抽样
df.sample(3)
```
