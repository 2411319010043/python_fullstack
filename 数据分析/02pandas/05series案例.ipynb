```python
'''创建一个包含10名学生数学成绩的series，成绩范围在50-100之间
    计算平均分、最高分、最低分、并找出高于平均分的学生人数'''
```

```python
np.random.seed(0)
scores = pd.Series(np.random.randint(50,101,10))
print(scores)
```

```python
print('平均分：',scores.mean())
print('最高分：',scores.max())
print('最低分：',scores.min())
print('高于平均分的学生人数：',len(scores[scores>scores.mean()]))
print('高于平均分的学生人数：',scores[scores>scores.mean()].count())
```

```python
'''给定某城市一周每条最高温度Series 
    找出温度超过30度的天数
    计算平均温度
    将温度从高到低排序
    找出温度变化量最大的两天'''
```

```python
temperatures = pd.Series([28,31,29,32,30,27,33],index=['周一','周二','周三','周四','周五','周六','周日'])
print(temperatures[temperatures>30].count())
print(temperatures.mean())
print(temperatures.sort_values(ascending=False))
```

```python
#找出温度变化量最大的两天
# 28 31 29 32 30 27 33
# none 3 -2 3 -2 -3 6
t3 = temperatures.diff().abs()  # 获取当前数据与前一个的差值
max_change_idx = t3.idxmax()  # 周日
max_pos = list(temperatures.index).index(max_change_idx)  # 获取temp中的周日的索引
prev_day_idx = temperatures.index[max_pos-1] # 获取周六
print(prev_day_idx,max_change_idx)
```

```python
'''给定某股票连续10个交易日的收盘价Series
    计算每日收益率(当日收盘价/前日收盘价-1)
    找出收益率最高和最低的日期
    计算波动率(收益率的标准差)'''
```

```python
prices=pd.Series(
    [102.3,103.5,105.1,104.8,106.2,107.0,106.5,108.1,109.3,110.2],
    index=pd.date_range('2023-01-01',periods=10))

pct = prices.pct_change()  # 计算收益率方法
print(pct)
print(pct.idxmax())
print(pct.idxmin())
print(pct.std())
```

```python
'''某产品过去12月的销售量Series
    计算季度平均销量(每3个月一个季度)
    找出销量最高的月份
    计算月环比增长率
    找出连续增长超过两个月的月份'''
```

```python
sales = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220],
                  index=pd.date_range('2022-1-1',periods=12,freq='MS'))
sales.resample('QS').mean()  # 重新采样方法 QS代表季度
print(sales.idxmax())  # 销量最高的月份
print('月环比增长率：',sales.pct_change())
#找出连续增长超过两个月的月份 利用滑动窗口
a = sales.pct_change()
b = a>0
b[b.rolling(3).sum() == 3].keys()  # 设定滑动窗口的大小

```

```python
'''某商店每小时销售额Series
    按天重采样计算每日总销售额
    计算每条营业时间(8:00-22:00)和非营业时间的销售额比例
    找出销售额最高的3个小时'''
```

```python
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0,100,24),
                         index=pd.date_range('2025-01-01',periods=24,freq='H'))
print(hourly_sales)
```

```python
# 按天重采样计算每日总销售额
day_sales = hourly_sales.resample('D').sum()
```

```python

hourly_sales.between_time('8:00','22:00')  # 筛选一段时间内的series
business_hours_sales = hourly_sales[(hourly_sales.index.hour>=8) & (hourly_sales.index.hour<=22)].sum()  # 获取索引中的小时
# 计算每条营业时间(8:00-22:00)和非营业时间的销售额比例
print(business_hours_sales/(day_sales - business_hours_sales))
```

```python
#找出销售额最高的3个小时
print(hourly_sales.nlargest(3))  # .nlargest(n) 是 pandas Series 和 DataFrame 的方法，用于返回前 n 个最大的值。
```
