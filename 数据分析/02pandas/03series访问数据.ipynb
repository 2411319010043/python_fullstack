```python
import pandas as pd
```

```python
s = pd.Series([10,2,3,4,5],index=['a','b','c','d','e'])
print(s)
```

```python
# 1. 直接访问
# print(s[1])  # 不推荐
print(s['a'])
```

```python
# 2. 布尔索引
print(s[s<3])
```

```python
# 访问头部
s['f'] = 6
print(s.head(2))  # 默认返回前5条数据
```

```python
# 访问结尾
print(s.tail(1))  # 默认返回后5行
```
