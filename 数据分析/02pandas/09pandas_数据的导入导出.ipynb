```python
# 缺失值的处理
# nan: not a number
import pandas as pd
import numpy as np
s = pd.Series([1,2,np.nan,None,pd.NA])
s
df = pd.DataFrame([[1,2,3,np.nan],[4,5,6,None],[7,8,9,33]])
df
```

```python
# 查看是否是缺失值
print(s.isna())
print(s.isnull())
```

```python
# 剔除缺失值
print(s.dropna())  # 只删除单个缺失值
print(df.dropna())  # 只要有缺失值就会删除一整行
print(df.dropna(how = 'all'))  # 若一行中所有的值都是缺失值，才删除一整行数据
print(df.dropna(thresh=2))  # 若至少有n个值不是缺失值，就保留
print(df.dropna(axis=1))  # 剔除一整列
print(df.dropna(subset=[3]))  # 指定检查某一列的缺失值 如果有就删除一整行
```

```python
# 填充缺失值
df = pd.read_csv('data/weather_withna.csv')
df.tail()
df.isna().sum(axis=0)
print(df.fillna({'temp_max':20,'wind':2.5}).tail()) #使用字典指定列填充固定值
print(df.fillna(df[['wind']].mean()).tail())  # 使用统计值来填充
print(df.ffill().tail())  # 使用前面的值相邻值填充
print(df.bfill().tail())  # 使用后面的值相邻值填充
```
