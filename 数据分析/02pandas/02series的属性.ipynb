```python
import pandas as pd
```

```python
s = pd.Series([10,2,3,4,5],index=['a','b','c','d','e'])
print(s)
```

```python
# 1. index 索引
print(s.index)
```

```python
# 2. values 值
print(s.values)
```

```python
# 3. shape ndim size
print(s.shape,s.ndim,s.size)
```

```python
# 4. dtype 数据类型 name 标签
s.name = 'test'
print(s.dtype,s.name)
```

```python
# loc[] 显式索引  iloc[] 隐式索引
print(s.loc['a'])
print(s.iloc[0])
print(s.loc['a':'c'])  # 索引切片
```

```python
# at[] 显式获取单个数值 iat[] 隐式获取单个元素 不支持切片
print(s.at['a'])
print(s.iat[0])
```
