```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([10,2,np.nan,None,3,4,5],index=['a','b','c','d','e','f','g'])
print(s)
```

```python
s.head(2)  # 默认取前5行的数据
s.tail(3)  # 默认取后5行的数据
```

```python
# 查看所有的描述性信息
s.describe()
```

```python
# 获取元素个数(忽略缺失值)
s.count()
```

```python
# 获取索引
print(s.keys())  # 方法
print(s.index)  # 属性
```

```python
# 判断是否是缺失值
s.isna()
```

```python
# 判断数值是否包含在数组中
s.isin([0])
```

```python
# 平均数
print(s.mean())
```

```python
# 标准差
print(s.std())
```

```python
# 方差
print(s.var())
```

```python
# 总和
print(s.sum())
```

```python
# 最大值、最小值
print(s.min(),s.max())
```

```python
# 中位数
print(s.median())
```

```python
# 排序
s.sort_values()
```

```python
# 分位数
print(s.quantile(0.75))
```

```python
# 众数
s['h'] = 4
print(s.mode())
```

```python
# 计算数据出现的次数
print(s.value_counts())
```

```python
# 去重
print(s.drop_duplicates())  # 返回series
print(s.unique())  # 返回list
print(s.nunique()) # 返回去重后的个数
```

```python
# 排序--值 索引
s.sort_index()
s.sort_values()
```
