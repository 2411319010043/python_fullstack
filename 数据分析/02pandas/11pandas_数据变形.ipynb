```python
import pandas as pd
```

```python
data = {
    'ID':[1,2],
    'name':['alice','bob'],
    'math':[90,85],
    'english':[88,92],
    'science':[95,89]
}
df = pd.DataFrame(data)
```

```python
# 1. 转置
df.T
```

```python
df  # 宽表
# 宽表变长表
df2 = pd.melt(df,id_vars=['ID','name'],var_name='科目',value_name='分数')
df2.sort_values('name')
```

```python
# 长表变宽表
pd.pivot(df2,index=['ID','name'],columns='科目',values='分数')
```

```python
data = {
    'ID':[1,2],
    'name':['alice smith','bob smith'],
    'math':[90,85],
    'english':[88,92],
    'science':[95,89]
}
df = pd.DataFrame(data)
```

```python
# 分列 (把一个列分成两列)
df[['first','last']] = df['name'].str.split(" ",expand=True)
df
```

```python
# 分列高压和低压
df = pd.read_csv('data/sleep.csv')
df = df[['person_id','blood_pressure']]
df[['high_blood_pressure','short_blood_pressure']] = df['blood_pressure'].str.split('/',expand=True)
df['high_blood_pressure'] = df['high_blood_pressure'].astype('int64')
df['short_blood_pressure'] = df['short_blood_pressure'].astype('int64')
df
```
