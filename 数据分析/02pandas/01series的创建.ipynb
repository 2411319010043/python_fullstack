```python
import pandas as pd
```

```python
# 创建series
s = pd.Series([10,2,3,4,5])
print(s)
```

```python
# 自定义索引
s = pd.Series([10,2,3,4,5],index=['a','b','c','d','e'])
print(s)
```

```python
# 定义标签name
s = pd.Series([10,2,3,4,5],index=['a','b','c','d','e'],name='月份')
print(s)
```

```python
# 通过字典创建
s = pd.Series({'a':1,'b':2,'c':3,'e':4})
print(s)
s1 = pd.Series(s,index=['a','c'])  # 通过已有的Series创建一个新的
print(s1)
```
