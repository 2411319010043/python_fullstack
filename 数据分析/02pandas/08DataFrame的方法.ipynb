```python
'''学生成绩分析
    某班级的学生成绩数据如下，
    1.计算每位学生的总分和平均分
    2.找出数学成绩高于90或英语成绩高于85分的学生
    3.按总分从高到低排序，并输出前三名学生'''
```

```python
data = pd.DataFrame({
    '姓名':['张三','李四','王五','赵六','钱七'],
    '数学':[85,92,78,88,95],
    '英语':[90,88,85,92,80],
    '物理':[75,80,88,85,90]
                     })
print(data)
```

```python
# 1.计算每位学生的总分和平均分 求某一行的总和
# print(df['score'].sum())  # 求某一列的和
data['总分'] = data[['数学','英语','物理']].sum(axis=1)
data['平均分'] = data['总分'] / 3
print(data)
```

```python
# 2.找出数学成绩高于90或英语成绩高于85分的学生
data[(data['数学'] > 90) | (data['英语'] > 85)]
```

```python
# 3.按总分从高到低排序，并输出前三名学生
print(data.sort_values(by='总分',ascending=False).head(3))
```

```python
'''销售数据分析
    某公司销售数据如下
    1.计算每种产品的总销售额(销售额 = 单价 * 销量)
    2.找出销售额最高的产品
    3.按销售额从高到低排序，并输出所有产品信息'''
```

```python
data = pd.DataFrame({
    '产品名称':['A','B','C','D'],
    '单价':[100,150,200,120],
    '销量':[50,30,20,40]
})
print(data)
```

```python
# 1.计算每种产品的总销售额(销售额 = 单价 * 销量)
data['总销售额'] = data['单价'] * data['销量']
print(data)
```

```python
# 2.找出销售额最高的产品
print(data.sort_values('总销售额',ascending=False).head(1))
```

```python
# 3.按销售额从高到低排序，并输出所有产品信息
print(data.sort_values('总销售额',ascending=False))
```

```python
'''电商用户行为分析
    1.计算每位用户的总消费金额(消费金额 = 商品单价 * 购买数量)
    2.找出消费金额最高的用户，并输出所有信息
    3.计算所有用户的平均消费金额(保留2位小数)
    4.统计电子产品的总购买数量'''
```

```python
data = pd.DataFrame({
    '用户ID':[101,102,103,104,105],
    '用户名':['Alice','Bob','Charlie','David','Eve'],
    '商品类别':['电子产品','服饰','电子产品','家居','服饰'],
    '商品单价':[1200,300,800,150,200],
    '购买数量':[1,3,2,5,4]
})
print(data)
```

```python
# 1.计算每位用户的总消费金额(消费金额 = 商品单价 * 购买数量)
data['总消费金额'] = data['商品单价'] * data['购买数量']
print(data)
```

```python
# 2.找出消费金额最高的用户，并输出所有信息
print(data.nlargest(1,columns=['总消费金额']))
```

```python
# 3.计算所有用户的平均消费金额(保留2位小数)
data['平均消费金额'] = data['总消费金额'] / data['购买数量']
print(data)
```

```python
# 4.统计电子产品的总购买数量
print(data[data['商品类别'] == '电子产品']['购买数量'].sum())
```
