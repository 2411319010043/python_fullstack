```python
import pandas as pd
import numpy as np
```

```python
# 通过Series来创建
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([6,7,8,9,10])
df = pd.DataFrame({'第1列':s1,'第2列':s2})
print(df)
print('df的类型：',type(df))
print('df里面s1的类型：',type(df['第1列']))
```

```python
# 通过字典来创建
df = pd.DataFrame({
    'id':[1,2,3,4,5],
    'name':['tom','jack','alice','luce','mike'],
    'age':[15,17,20,26,30]
},index=[1,2,3,4,5],columns=['name','id','age']  # columns调整列的顺序
)
print(df)
```
