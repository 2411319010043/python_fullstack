```python
import numpy as np
```

### 1.ndarray的属性

```python
arr = np.array(1)
print(arr)
print('数组的形状：', arr.shape)
```

```python
arr = np.array([1,2.5,3])
print(arr)
print('数组的形状：', arr.shape)
print('数组的维度:', arr.ndim)
print('元素的个数:', arr.size)
print('元素的数据类型:', arr.dtype)
print('元素的转置', arr.T)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print('数组的形状：', arr.shape)
print('数组的维度:', arr.ndim)
print('元素的个数:', arr.size)
print('元素的数据类型:', arr.dtype)
print('元素的转置', arr.T)
```
