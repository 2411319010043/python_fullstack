```python
import numpy as np
```

### 1. ndarray的创建

```python
# 基础的创建方法
list1 = [4,5,6]
arr = np.array(list1, dtype= float)  # 方法/函数 指定数据类型
print(arr.ndim)  # 属性
print(arr)
```

```python
# copy
arr1 = np.copy(arr)  # 元素和原始数据相同，但不是一个数组
print(arr1)
arr1[0] = 8  # 不是指向同一内存地址
print(arr1)
print(arr)
```

```python
# 预定义形状
# 全0 全1 未初始化 固定值
# 1. 全0
arr = np.zeros((2,3))  # zeros方法默认是float64
print(arr)

arr = np.zeros((2,))  # 1维
print(arr)
```

```python
# 2. 全1
arr = np.ones((2,))  
print(arr)
```

```python
# 3. 未初始化
arr = np.empty((4,3))
print(arr)
```

```python
# 4. 固定值
arr = np.full((3,4),2025)
print(arr)
```

```python
arr1 = np.zeros_like(arr) # 像arr数组一样的0数组
print(arr1)
arr1 = np.empty_like(arr)
print(arr1)
arr1 = np.ones_like(arr)
print(arr1)
arr1 = np.full_like(arr,2026)
print(arr1)
```

```python
# 等差数列 1 2 3 4 5
arr = np.arange(2,11,2)  # start,end,step
print(arr)
```

```python
# 等间隔数列
arr = np.linspace(0,10,5) # 把数据均分成5份
print(arr)
```

```python
# 练习 用arange实现
arr = np.linspace(0, 100, 5, dtype=int) # 把数据均分成5份
print(arr)
arr = np.arange(0,101,25)
print(arr)
```

```python
# 对数间隔数列
arr = np.logspace(0,4,3,base=2) 
print(arr)
```

#### 特殊矩阵的构造

```python
# 单位矩阵：主对角线上的元素为1，其他为0
arr = np.eye(3, 4, dtype=int)  
print(arr)
```

```python
# 对角矩阵：主对角线上是非零的元素，其他为0
arr = np.diag([5,1,2,3])
print(arr)
```

#### 随机数组的生成

```python
# 生成0-1之间的随机浮点数(均匀分布)
arr = np.random.rand(2,3)
print(arr)
```

```python
# 生成指定范围之间的随机浮点数(均匀分布)
arr = np.random.uniform(3,6,(2,3))
print(arr)
```

```python
# 生成指定范围之间的随机整数
arr = np.random.randint(3,6,(2,3))
print(arr)
```

```python
# 生成随机数量(正态分布:两边小，中间大) (-3到3之间)
arr = np.random.randn(2,3)
print(arr)
```

```python
# 设置随机种子
np.random.seed(20)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
