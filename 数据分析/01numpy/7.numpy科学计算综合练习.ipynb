```python
import numpy as np
```

```python
'''某城市一周的最高气温为[28,30,29,31,32,30,29]
    计算平均气温，最高气温和最低气温
    气温超过30度的天数'''
```

```python
temps = np.array([28,30,29,31,32,30,29])
print('平均气温:%.3f' %np.mean(temps))
print('最高气温:%i' %np.max(temps))
print('最低气温:%i' %np.min(temps))
print('气温超过30的天数:',len(temps[temps>30]))
print(np.cumsum(np.where(temps>30,1,0))[-1])
print(np.count_nonzero(temps>30))
```

```python
'''学生成绩统计
    某班级5名学生数学成绩为[85,90,78,92,88]
    计算成绩的平均分、中位数和标准差
    将成绩转换成百分制(满分10)'''
```

```python
score = np.array([85,90,78,92,88])
print('平均分：%.3f'%np.mean(score))
print('中位数: %.3f'%np.median(score))
print('标准差: %.3f'%np.var(score))
print(score/10)
```

```python
'''矩阵运算
给定矩阵 A = [[1,2],[3,4]] 和 B = [[5,6], [7,8]]
计算 A + B 和 A * B(逐元素乘法)
计算 A和B的矩阵乘法(点积)'''
```

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])
print(a+b)
print(a*b)

print(a@b)
print(np.dot(a,b))
```

```python
'''随机数据生成
    生成一个(3,4)的随机整数数组，范围[0,10)
    计算每列的最大值和每行的最小值
    将数组中所有奇数替换为-1'''
```

```python
np.random.seed(0)
arr = np.random.randint(0,10,(3,4))
print(arr)
print(np.max(arr,axis=0))  # axis表示轴 0是列 1是行
print(np.min(arr,axis=1))
#print(np.where(arr%2==0, arr, -1))
arr[arr%2==1]=-1
print(arr)
```

```python
'''数组变形
    创建一个1-12的一维数组，并转换为(3，4)的二维数组
    计算每行的和与每列的平均值
    将数组展平为一维数组'''
```

```python
arr = np.arange(1,13,1)
print(arr)
arr1 = np.reshape(arr,(3,4))
print(arr1)
print('每行的和：',np.sum(arr1,axis=1))
print('每列的平均值',np.mean(arr1,axis=0))
print(arr.flatten())
```

```python
'''布尔索引
    生成一个(5,5)的随机数组，范围[0,20)
    找出数组中大于10的元素
    将所有大于10的元素替换为0
'''
```

```python
np.random.seed(0)
arr = np.random.randint(0,20,(5,5))
print(arr)
print(arr[arr>10])
print(np.where(arr>10,0,arr))
```

```python
'''统计函数应用
    某公司6个月的销售额(万元)为[120,135,110,125,130,140]
    计算销售额的总和、均值和方差
    找出销售额最高的月份和最低的月份'''
```

```python
arr = np.array([120,135,110,125,130,140])
print('销售额总和：',np.sum(arr))
print('均值：',np.mean(arr))
print('方差：',np.var(arr))
print('销售额最高的月份：',np.argmax(arr)+1)
print('销售额最低的月份：',np.argmin(arr)+1)
```

```python
'''数组拼接
    给定a = [1,2,3] 和 b=[4,5,6]
    将a和b水平拼接为一个新数组
    将a和b垂直拼接为一个新数组'''
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.concatenate([a,b])  # 拼接数组
print(c)
print(np.reshape(c,(2,3)))
```

```python
'''唯一值与排序
    给定数组[2,1,2,3,1,4,3]
    找出数组中的唯一值并排序
    计算每个唯一值出现的次数'''
```

```python
arr = np.array([2,1,2,3,1,4,3])
u_arr,counts = np.unique(arr,return_counts=True)
print(u_arr,counts)
```

```python
'''综合应用
    某商店5天的销售额(万元)和成本(万元)如下：
    销售额：[20,25,22,30,28]
    成本：[15,18,16,22,20]
    计算每天的利润(销售额-成本)
    计算利润的平均值和标准差
    找出利润最高的天数'''
```

```python
money = np.array([20,25,22,30,28])
basic = np.array([15,18,16,22,20])
lirun = money - basic
print(lirun)
print('平均值：',np.mean(lirun))
print('标准差:',np.std(lirun))
print( len(lirun[lirun==np.max(lirun)] ))
```
